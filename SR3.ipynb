```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'C:/Users/vuyis/Downloads/Closed Ended Data.csv'
data = pd.read_csv('C:/Users/vuyis/Downloads/Closed Ended Data.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='EDU_LEVEL', hue='EDU_LEVEL', palette='Set2', legend=False)
plt.title('Distribution of Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Histogram for Numeric Variables
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='AVG_MON_SAL', bins=20, kde=True)
plt.title('Distribution of Average Monthly Salary')
plt.xlabel('Average Monthly Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Check the column names in the dataset
print(data.columns)
```

```python
plt.figure(figsize=(8 , 8))
data['LEADER_PARTICIPATION'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Leader Participation Proportion')
plt.show()
```

```python
# Example: Box plot of average monthly salary
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, y='AVG_MON_SAL', x='AVG_MON_SAL', palette='pastel', hue='AVG_MON_SAL', dodge=False)
plt.title('Distribution of Average Monthly Salary')
plt.ylabel('Average Monthly Salary')
plt.xlabel('Average Monthly Salary')
plt.legend([],[], frameon=False)  # Remove the legend
plt.show()
```

```python
plt.figure(figsize=(10 ,6))
sns.countplot(data= data,x ='AVG_MON_SAL')
plt.title('Distribution of Average Monthly Salary')
plt.ylabel('Average Monthly Salary')
plt.xlabel('Average Monthly Salary')
plt.show()
```
